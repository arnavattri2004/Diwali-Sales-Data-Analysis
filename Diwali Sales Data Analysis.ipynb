# Exploratory Data Analysis(EDA)

### Gender

```python
ax = sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y="Amount", data = sales_gen, hue='Gender')
```

From the above graphs, we can see that most of the buyers are females and purchasing power of females is also more than the males.

```python
df.columns
```

## Age

```python
ax = sns.countplot(x='Age Group', data = df, hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y="Amount", data = sales_age, hue='Age Group')
```

From the above graph, we can see that the most buyers are Females between the Age Group of 26-35.  

## State

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='State', y="Orders", data = sales_state, hue='State')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='State', y="Amount", data = sales_state, hue='State')
```

From the above graph, we can see that most of the order and total sales/amount is from Uttar Pradesh, Maharashtra and Karnataka respectively.

## Marital Status

```python
ax = sns.countplot(x='Marital_Status', data = df)
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ms = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Marital_Status', y="Amount", data = sales_ms, hue='Gender')
```

From these graphs, we ccan see that most of the buyers are married women and they have higher purchasing power.

## Occupation

```python
ax = sns.countplot(x='Occupation', data = df, hue='Occupation')
sns.set(rc={'figure.figsize':(25,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(x='Occupation', y="Amount", data = sales_occ, hue='Occupation')
```

From the above graph, we can see that most of the buyers are from the IT, Healthcare and Aviation Sector.

## Product Category

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = df, x = 'Product_Category', hue='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_prod_cat = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_prod_cat, x = 'Product_Category',y= 'Amount', hue='Product_Category')
```

From the above graphs, we can see that most of the sold products are from Food, Clothing and Electronics Category.

## Product ID

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders', hue='Product_ID')
```

# Conclusion

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation sector and are more likely to buy products from Food, Clothing and Electronics category.
